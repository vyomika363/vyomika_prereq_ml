**Question1**
between a n*k and k*m matrix,
when they are multiplied there will be nm elements, out of which eack one will have k multiplications, so total multiplications is k*n*m.

For additions, n*m elements will have (k-1) additions, so total (k-1)*m*n

```python
#Question2(using line)
def multiplication(A,B):
    n= len(A)
    k=len(A[0])
    m=len(B[0])
    mat=[]
    for i in range(n):
      row=[]
      for j in range(m):
          row.append(0)
      mat.append(row)

    for i in range(n):
        for j in range(m):
            sum=0
            for l in range(k):
                sum+=A[i][l]*B[l][j]
            mat[i][j]=sum
    return mat
A=[[1, 2],
   [2, 3]]
B=[[4,2],
   [1,1]]
print(multiplication(A,B))
```

```python
#Question2 (using numpy)

import numpy as np
def multip(A,B):
    mat=np.dot(A,B)
    return mat
A=[[1, 2],
   [2, 3]]
B=[[4,2],
   [1,1]]
print(multip(A,B))
```

```python
#Question 6
#gradient using sympy
import sympy as sp
import numpy as np

sp.init_printing()
x,y = sp.symbols('x y')

def gradient(f):
    return (f.diff(x), f.diff(y))

f = x**2*y + sp.sin(x) *y**3
g = gradient(f)
print(g)
```

```python
#Question 10

import numpy as np
import pandas as pd
mat=np.random.uniform(1,2,size=(20,5))
columns=pd.DataFrame(mat, columns=['a','b', 'c', 'd', 'e'])

def mean(A):
    val=0
    for i in range(len(A)):
        val+=A[i]
    return val/len(A)
def dev(A):
    return (A-mean(A))
def var(A):
    stan_dev=dev(A)**2
    return mean(stan_dev)
lst=[var(columns['a']), var(columns['b']), var(columns['c']), var(columns['d']), var(columns['e'])]
sort_list=sorted(lst)
print(sort_list[-1])

lst_1=[]
for i in range(20):
    row=[columns['a'][i], columns['b'][i], columns['c'][i], columns['d'][i], columns['e'][i]]
    lst_1.append(mean(row))
lst_sort=sorted(lst_1)
print(lst_sort[0])
```

```python
#Question 11
import numpy as np
import pandas as pd
mat=np.random.uniform(1,2,size=(20,5))
print(mat)
columns=pd.DataFrame(mat, columns=['a','b', 'c', 'd', 'e'])
f=[]
for i in range(20):
    f.append(columns['a'][i]+columns['b'][i]+columns['c'][i]+columns['d'][i]+columns['e'][i])
g=[]
count=0
val_LT8=[]
val_GT8=[]
for j in range(20):
    if f[j]<8:
        g.append("LT8")
        count+=1
        val_LT8.append(j)
    else:
        g.append("GT8")
        val_GT8.append(j)
print(count)
val=0
for i in val_LT8:
    val+=f[i]
mean_LT8_f=val/len(val_LT8)
lst1=(f-mean_LT8_f)
stan_dev=lst1**2
valu=0
for j in val_LT8:
    valu+=stan_dev[j]
ans1=valu/len(val_LT8)
print(ans1)

val2=0
for i in val_GT8:
    val2+=f[i]
mean_GT8_f=val2/len(val_GT8)
lst2=(f-mean_GT8_f)
stan_dev2=lst2**2
valu2=0
for k in val_GT8:
    valu2+=stan_dev2[k]
ans2=valu2/len(val_GT8)
print(ans2)
```
